# Loading the data

```python
DATA = "2016Census_G59_AUS.csv"
```

```python
with open(DATA) as file:
    contents = file.read()
    print(contents)
    
```

# Data Inspection

```python
lines = contents.splitlines()
lines
```

```python
headings = lines[0].split(',')
headings[:10]
```

```python
numbers = lines[1].split(',')
numbers[:10]
```

```python
print(len(headings), len(numbers))
```

## Data Cleaning & Conversion

```python
headings = headings[1:]
numbers = numbers[1:]
print(len(headings), len(numbers))
```

```python
print(headings[:5])
print(numbers[:5])

print(type(headings))
print(type(numbers))
```

```python
num_people = []
for stringdata in numbers:
    num_people.append(int(stringdata))
print(type(num_people[0]))
```

## Importing Package

```python
import matplotlib.pyplot as plt

```

```python
plt.bar(headings, num_people)
```

```python
headings = headings[:-3]
num_people = num_people[:-3]
len(headings)
```

```python
plt.bar(headings, num_people)
plt.show()
```

```python
headings[2::3]
```

```python
wanted = slice(2,len(num_people),3)


travel_headings = headings[wanted]
travel_people = num_people[wanted]

print(len(travel_headings), len(travel_people))
```

```python
plt.bar(travel_headings, travel_people)
plt.show()
```

```python
plt.figure(figsize= (9,4))
plt.bar(travel_headings, travel_people)
plt.tick_params(axis = 'x', labelrotation = 90)
plt.show()
```

```python
ix = travel_headings.index("One_method_Tot_one_method_P")
num = travel_people.pop(ix)
```

```python
travel_headings.pop(ix)
```

```python
ix = travel_headings.index("Two_methods_Tot_two_methods_P")
travel_headings.pop(ix)
travel_people.pop(ix)
```

```python
travel_headings.pop(ix)
travel_people.pop(ix)
```

```python
print(len(travel_headings), len(travel_people))
```

```python
travel_categories = []
travellers = []

for i in range(len(travel_headings)):
    if "Tot_" not in travel_headings[i]:
        travel_categories.append(travel_headings[i][:-2])
        travellers.append(int(travel_people[i]))
```

```python
print(len(travel_categories), len(travellers))
```

```python
plt.figure(figsize= (9,4))
plt.bar(travel_categories, travellers)
plt.tick_params(axis = 'x', labelrotation = 90)
plt.xlabel("Methods of Transport")
plt.ylabel("Number of users")
plt.title("Methods of Travel to Work, 2016 Census")
plt.show()
```

```python
plt.pie(travellers, labels= travel_categories, rotatelabels= True)
plt.title("Method of Travel")
plt.show()
```

- sort

```python
pairs =[]
for i in range(len(travellers)):
    pairs.append((travellers[i], travel_categories[i]))
pairs.sort(reverse = True)
    
```

```python
sorted_travel_cats = []

sorted_travellers = []

for i in range(len(pairs)):
    sorted_travel_cats.append(pairs[i][1])
    sorted_travellers.append(pairs[i][0])
sorted_travel_cats[:5]
sorted_travellers[:5]
```

```python
plt.figure(figsize= (14,6))
plt.pie(sorted_travellers, labels= sorted_travel_cats, rotatelabels= True)
plt.title("Method of Travel")
plt.show()
```

- Using Zip

```python

pairs = list(zip(travellers, travel_categories))
pairs.sort(reverse= True)
```

```python
from operator import itemgetter

```

```python

pairs = list(zip(travel_categories, travellers))
pairs
pairs.sort(key = itemgetter(1),reverse= True)
print(pairs[:5])
```

- Grouping together others

```python
PIECES_OF_PIE = 9
```

```python
plt.figure(figsize= (14,6))
others = sum(sorted_travellers[PIECES_OF_PIE-1:])
plt.pie(sorted_travellers[:PIECES_OF_PIE-1] + [others], labels= sorted_travel_cats[:PIECES_OF_PIE-1] + ["others methods of transport"])
plt.savefig('transport.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize= (9,4))
plt.semilogy()
plt.bar(sorted_travel_cats, sorted_travellers)
plt.grid(axis = "both", which = "both", alpha = 0.6)
plt.tick_params(axis = 'x', labelrotation = 90)
plt.xlabel("Methods of Transport")
plt.ylabel("Number of users")
plt.title("Methods of Travel to Work, 2016 Census")
plt.show()
plt.savefig("Methods of travel to work.jpg",dpi = 300)
```
