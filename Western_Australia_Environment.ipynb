```python
DATA_LENGTH = 1476

dates = np.zeros(DATA_LENGTH, dtype = np.float32)
msls = np.zeros(DATA_LENGTH,dtype = np.int32)
missing_days = np.zeros(DATA_LENGTH,dtype = np.uint8)
flags = np.zeros_like(missing_days)


index = 0
with open(DATA) as datafile:
    for line in datafile:
        linedata = line.split(";")
        dates[index] = linedata[0]
        msls[index] = linedata[1]
        missing_days[index] = linedata[2]
        flags[index] = linedata[3]
        index = index +1
        
print(dates[:5])
print(msls[:5])
print(missing_days[:5])
print(flags[:5])
        

```

# repeat with enumerate

```python
DATA_LENGTH = 1476

a_dates = np.zeros(DATA_LENGTH)
a_msls = np.zeros(DATA_LENGTH,dtype = int)
a_missing_days = np.zeros_like(a_msls)
a_flags = np.zeros_like(a_msls)


with open(DATA) as datafile:
    for (index, line) in enumerate(datafile):
        linedata = line.split(";")
        a_dates[index] = linedata[0]
        a_msls[index] = linedata[1]
        a_missing_days[index] = linedata[2]
        a_flags[index] = linedata[3]

        
print(a_dates[:5])
print(a_msls[:5])
print(a_missing_days[:5])
print(a_flags[:5])
        
```

- b) [:-1]

```python
xs = np.linspace(-np.pi,np.pi)
sines = np.sin(xs)
cosines = np.cos(xs)

plt.plot(xs,sines,linewidth = 2)
plt.plot(xs,cosines, linewidth = 2)
plt.show()
```

```python
xs = np.linspace(-np.pi,np.pi)
sines = np.sin(xs)
cosines = np.cos(xs)

plt.scatter(xs,sines,linewidth = 2)
plt.scatter(xs,cosines, linewidth = 2)
plt.show()
```

```python
xs = np.linspace(-1,1)
sines = np.sin(xs*np.pi)
cosines = np.cos(xs*np.pi)

plt.plot(xs,sines,linewidth = 2, label = "sine")
plt.plot(xs,cosines, linewidth = 2, label = "cosine")
plt.xlabel('\u03a0')
plt.legend()
plt.title("Looks like a phase shift")
plt.show()
```
